```python
#Напишите функцию summa_n, которая принимает одно целое положительное число t и находит сумму всех чисел от 1 до t включительно.
#Вызывать в скрипте функцию не нужно.
def summa_n(t):
    t = (int(input("Введите число: ")))
    l1 = [t]
    for i in range(1,t): #сама ф-ция
        l1.append(i)
    S = sum(l1)
    return f"Я знаю, что сумма чисел от 1 до {t} равна {S}"
```

Напишите функцию area_and_length(), определяющую площадь круга и длину окружности, по заданному радиусу R. Используйте модуль math

! Результат округлите до сотых

Вывод результата в виде строки (без кавычек):

"Circle area = circle_area, circle length = circle_length"
Файл с решением должен иметь название task5_2.py, название функции - def area_and_length(). Вызывать в скрипте функцию не нужно.

Тестовый пример:

area_and_length(23)

Вывод:
Circle area = 1661.9, circle length = 144.51

```python
import math
def area_and_length(stdin):
    stdin = float()
    from math import pi
    circle_area = pi*pow(stdin,2)
    circle_area = round(circle_area,2)
    circle_lenght = pi*(stdin*2)
    circle_lenght = round(circle_lenght, 2)
    return f'Circle area = {circle_area}, circle length = {circle_lenght}'
```

```python
def read_file_last(data,n):
  file = open(data, "r+")
  str_list = file.readlines()
  new_list = [i.replace("\n,"") for i in str_list]
  file.close()
  return new_list[-n:]
```
